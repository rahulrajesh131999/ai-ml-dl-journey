> # Introduction

# Ordinary Least Squares (OLS) — A Visual Walkthrough

> *How do we mathematically find the best straight-line relationship between inputs (features) and outputs (targets)?*

---

## 1. The Core Idea of Linear Regression

Suppose you want to predict house prices. Your features might be: size, number of rooms, age of house.

A linear model assumes the prediction is a **weighted sum of features**:

$$\hat{y} = w_0 + w_1x_1 + w_2x_2 + \dots + w_px_p$$

Where:
- $x_i$ = feature values
- $w_i$ = coefficients/weights
- $\hat{y}$ = predicted output

**Example:**

$$\text{price} = 50000 + 200(\text{size}) + 10000(\text{rooms})$$

- Every extra square foot adds \$200
- Every extra room adds \$10,000

---

## 2. Why Add the Fake Feature ($x_0 = 1$)?

We rewrite the model by defining $x_0 = 1$, so:

$$\hat{y} = \sum_{j=0}^{p} w_j x_j$$

This is just **mathematical convenience** — the bias/intercept $w_0$ becomes part of the same unified formula, making matrix algebra much cleaner:

$$\underbrace{1 \cdot w_0}_{\text{bias}} + x_1 w_1 + x_2 w_2$$

---

## 3. Matrix Form

For many observations, we stack everything into matrices.

| size | rooms | price   |
|------|-------|---------|
| 1000 | 2     | 200,000 |
| 1500 | 3     | 300,000 |
| 2000 | 4     | 400,000 |

After adding the fake column of 1s, we get the **design matrix** $X$:

$$X = \begin{bmatrix} 1 & 1000 & 2 \\ 1 & 1500 & 3 \\ 1 & 2000 & 4 \end{bmatrix}, \quad w = \begin{bmatrix} w_0 \\ w_1 \\ w_2 \end{bmatrix}, \quad y = \begin{bmatrix} 200000 \\ 300000 \\ 400000 \end{bmatrix}$$

The model becomes the **central equation of linear regression**:

$$\boxed{y = Xw + \varepsilon}$$

---

## 4. The Error Term $\varepsilon$

Real data is noisy. Even two houses with identical features may differ in price due to location, construction quality, or hidden factors.

$$\text{actual value} = \underbrace{Xw}_{\text{systematic linear part}} + \underbrace{\varepsilon}_{\text{random noise}}$$

---

## 5. Gauss-Markov Assumptions

These assumptions are what make OLS mathematically "good".

### Assumption 1: Zero Mean Errors
$$E[\varepsilon] = 0$$
Errors average out to zero — the model is not systematically over- or under-predicting.

### Assumption 2: Homoscedasticity (Equal Variance)
$$\text{Var}(\varepsilon_i) = \sigma^2$$
The noise level is roughly constant everywhere. Violations occur when, e.g., expensive houses have much larger errors than cheap ones.

### Assumption 3: Uncorrelated Errors
$$\text{Cov}(\varepsilon_i, \varepsilon_j) = 0 \quad \text{for } i \neq j$$
One error should not influence another — especially important in time series or sequential data.

---

## 6 & 7. Ordinary Least Squares (OLS)

We don't know the true weights $w$ — we only observe data. So we compute an **estimate** $\hat{w}$.

**OLS chooses weights that minimize the sum of squared prediction errors:**

$$\min_w \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$$

### Why square the errors?
1. Positive and negative errors don't cancel each other out
2. Large errors are penalized more heavily
3. The math becomes analytically solvable

---

## 8. Geometric Intuition

OLS finds the line/plane that gets as close as possible to **all data points simultaneously**:

| # Features | Geometry        |
|------------|-----------------|
| 1          | Best-fit line   |
| 2          | Best-fit plane  |
| $p$        | Best-fit hyperplane |

> **Core intuition:** Minimize total squared *vertical* distances from each data point to the fitted surface.

---

## 9. The Normal Equation (Closed-Form Solution)

Minimizing the least-squares loss via calculus gives the **normal equation**:

$$\boxed{\hat{w} = (X^TX)^{-1}X^Ty}$$

### Breaking it apart intuitively:

| Term | Meaning |
|------|---------|
| $X^TX$ | Relationships (correlations) between features |
| $(X^TX)^{-1}$ | Corrects for overlap/redundancy among features |
| $X^Ty$ | Relationships between features and targets |

Combining them gives the weight vector that **minimizes squared error**.

---

## 10 & 11. Why OLS is Special — The Gauss-Markov Theorem

An estimator is called **linear** if it can be written as:
$$\hat{w} = Ay$$
for some matrix $A$. OLS satisfies this since:
$$\hat{w} = \underbrace{(X^TX)^{-1}X^T}_{A} \, y$$

The **Gauss-Markov theorem** states:

> *Among all linear unbiased estimators, OLS has the minimum variance.*

This makes OLS the **BLUE** estimator:

| Letter | Meaning |
|--------|---------|
| **B** | Best (lowest variance) |
| **L** | Linear (in $y$) |
| **U** | Unbiased (correct on average) |
| **E** | Estimator |

---

## Summary — The Full Picture

$$\text{Data} \rightarrow X, y$$

$$\text{Model} \rightarrow y = Xw + \varepsilon$$

$$\text{Goal} \rightarrow \text{find } w$$

$$\text{Method} \rightarrow \text{minimize squared errors}$$

$$\text{Solution} \rightarrow \hat{w} = (X^TX)^{-1}X^Ty$$

$$\text{Theory} \rightarrow \text{Gauss-Markov says OLS is optimal}$$

---

**The one-line intuition:**
> *Linear regression finds the hyperplane that minimizes total squared vertical distances to the data points.*
